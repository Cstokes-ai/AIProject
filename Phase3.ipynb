```python
# Import necessary libraries
import pandas as pd
import tensorflow as tf
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt

# Reload the dataset
dataset = pd.read_csv('loan_data.csv')

# Identify categorical columns
categorical_columns = dataset.select_dtypes(include=['object']).columns

# Apply one-hot encoding to categorical columns
dataset = pd.get_dummies(dataset, columns=categorical_columns, drop_first=True)

# Shuffle the dataset
shuffle_dataset = dataset.sample(frac=1, random_state=42).reset_index(drop=True)

# Split the data into training and test sets
train_set = shuffle_dataset[:int(len(shuffle_dataset) * 0.8)]
test_set = shuffle_dataset[int(len(shuffle_dataset) * 0.8):]

# Normalize the numeric features
normalize_data = StandardScaler()
train_features = normalize_data.fit_transform(train_set.drop('loan_status', axis=1))
test_features = normalize_data.transform(test_set.drop('loan_status', axis=1))

# Extract labels
train_labels = train_set['loan_status']
test_labels = test_set['loan_status']

# Logistic Regression Model
logistic_model = LogisticRegression()
logistic_model.fit(train_features, train_labels)

# Evaluate Logistic Regression on the test set
logistic_predictions = logistic_model.predict(test_features)

# Calculate metrics for Logistic Regression
logistic_precision = precision_score(test_labels, logistic_predictions)
logistic_recall = recall_score(test_labels, logistic_predictions)
logistic_f1 = f1_score(test_labels, logistic_predictions)
logistic_accuracy = accuracy_score(test_labels, logistic_predictions)

# Neural Network Models (Different Architectures)
model_architectures = [
    (64, 32, 16, 8, 1),
    (32, 16, 8, 1),
    (16, 8, 1),
    (8, 1),
    (4, 1),
    (2, 1)
]

results = []

# Loop through different model architectures
for architecture in model_architectures:
    model = Sequential()
    
    # Add the layers based on the architecture
    model.add(Dense(architecture[0], activation='relu', input_shape=(train_features.shape[1],)))
    for units in architecture[1:-1]:
        model.add(Dense(units, activation='relu'))
    model.add(Dense(architecture[-1], activation='sigmoid'))
    
    # Compile and train the model
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    history = model.fit(train_features, train_labels, epochs=10, batch_size=32, validation_split=0.2, verbose=0)
    
    # Evaluate the model on the test set
    test_loss, test_accuracy = model.evaluate(test_features, test_labels, verbose=0)
    
    # Make predictions with the Neural Network model
    nn_predictions = (model.predict(test_features) > 0.5).astype("int32")
    
    # Calculate metrics for Neural Network
    nn_precision = precision_score(test_labels, nn_predictions)
    nn_recall = recall_score(test_labels, nn_predictions)
    nn_f1 = f1_score(test_labels, nn_predictions)
    
    # Append the results for each model
    results.append({
        "Model": f"Neural network model {architecture}",
        "Acc. on Training Set": f"{history.history['accuracy'][-1] * 100:.2f} %",
        "Acc. on Validation Set": f"{history.history['val_accuracy'][-1] * 100:.2f} %",
        "Precision": f"{nn_precision * 100:.2f} %",
        "Recall": f"{nn_recall * 100:.2f} %",
        "F1-score": f"{nn_f1:.4f}"
    })

# Logistic Regression Results
results.append({
    "Model": "Logistic Regression",
    "Acc. on Training Set": f"{logistic_accuracy * 100:.2f} %",
    "Acc. on Validation Set": "N/A",  # Logistic regression does not use validation set in this code
    "Precision": f"{logistic_precision * 100:.2f} %",
    "Recall": f"{logistic_recall * 100:.2f} %",
    "F1-score": f"{logistic_f1:.4f}"
})

# Convert results to DataFrame for display
results_table = pd.DataFrame(results)

# Display the results table
print("\nModel Comparison Results")
print(results_table)

# Plot learning curves for the best Neural Network model
plt.figure(figsize=(12, 6))
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Learning Curves for Neural Network')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
